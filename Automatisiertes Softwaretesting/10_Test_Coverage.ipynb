### Was bedeuten die Spalten?

| Spalte | Bedeutung | Beispiel |
|--------|-----------|----------|
| **Name** | Dateiname | `library.py` |
| **Stmts** | Anzahl Zeilen Code | 25 Zeilen |
| **Miss** | Nicht getestete Zeilen | 5 Zeilen fehlen |
| **Cover** | Prozentsatz | 80% getestet |

**Einfache Formel:**

```
Coverage = (Stmts - Miss) / Stmts Ã— 100%
         = (25 - 5) / 25 Ã— 100%
         = 80%
```

---

### Was ist gut?

| Coverage | Bewertung | Bedeutung |
|----------|-----------|----------|
| **< 50%** | ğŸ”´ Schlecht | Viel ungetesteter Code! |

| **> 90%** | ğŸŸ¢ Sehr gut | Exzellente Testabdeckung |

````python
# Unsere Library-Klasse (mit einigen Funktionen)

class BookAlreadyExistsError(Exception):
    """Wird ausgelÃ¶st, wenn ein Buch bereits existiert."""
    pass

class BookNotFoundError(Exception):
    """Wird ausgelÃ¶st, wenn ein Buch nicht gefunden wird."""
    pass


class Book:
    """ReprÃ¤sentiert ein Buch."""
    
    def __init__(self, isbn, title, author, year):
        self.isbn = isbn
        self.title = title
        self.author = author
        self.year = year
    
    def __str__(self):
        return f"{self.title} von {self.author} ({self.year})"


class Library:
    """Verwaltet eine Sammlung von BÃ¼chern."""
    
    def __init__(self):
        self.books = {}  # Zeile 1: Wird getestet? âœ…
    
    def add_book(self, book):
        """FÃ¼gt ein Buch hinzu."""
        if book.isbn in self.books:                    # Zeile 2: Wird getestet?
            raise BookAlreadyExistsError("Duplikat!")  # Zeile 3: Wird getestet?
        self.books[book.isbn] = book                   # Zeile 4: Wird getestet? âœ…
    
    def get_book(self, isbn):
        """Gibt ein Buch zurÃ¼ck."""
        if isbn not in self.books:                     # Zeile 5: Wird getestet?
            raise BookNotFoundError("Nicht gefunden!") # Zeile 6: Wird getestet?
        return self.books[isbn]                        # Zeile 7: Wird getestet? âœ…
    
    def count_books(self):
        """Gibt die Anzahl der BÃ¼cher zurÃ¼ck."""
        return len(self.books)  # Zeile 8: Wird getestet? âœ…
    
    def remove_book(self, isbn):
        """Entfernt ein Buch."""
        if isbn not in self.books:                     # Zeile 9: Wird getestet?
            raise BookNotFoundError("Nicht gefunden!") # Zeile 10: Wird getestet?
        del self.books[isbn]                           # Zeile 11: Wird getestet?


print("âœ… Library-Klasse definiert")
print("   Frage: Welche Zeilen werden von Tests ausgefÃ¼hrt?")
````

````python
# Unsere Tests (UNVOLLSTÃ„NDIG!)

def test_library_starts_empty():
    """Test: Neue Bibliothek ist leer."""
    library = Library()  # â† FÃ¼hrt Zeile 1 aus âœ…
    assert library.count_books() == 0  # â† FÃ¼hrt Zeile 8 aus âœ…


def test_add_book():
    """Test: Buch hinzufÃ¼gen."""
    library = Library()
    book = Book("978-3-16-148410-0", "Clean Code", "Robert C. Martin", 2008)
    library.add_book(book)  # â† FÃ¼hrt Zeile 2, 4 aus âœ…
    assert library.count_books() == 1


def test_get_book():
    """Test: Buch abrufen."""
    library = Library()
    book = Book("978-3-16-148410-0", "Clean Code", "Robert C. Martin", 2008)
    library.add_book(book)
    retrieved = library.get_book("978-3-16-148410-0")  # â† FÃ¼hrt Zeile 5, 7 aus âœ…
    assert retrieved == book


# Tests ausfÃ¼hren
print("ğŸ§ª Tests ausfÃ¼hren...\n")

test_library_starts_empty()
print("âœ… test_library_starts_empty PASSED")

test_add_book()
print("âœ… test_add_book PASSED")

test_get_book()
print("âœ… test_get_book PASSED")

print("\n" + "="*60)
print("COVERAGE-ANALYSE")
print("="*60)
print("\nğŸ“Š Welche Zeilen wurden ausgefÃ¼hrt?\n")
print("   âœ… Zeile 1: __init__ (self.books = {})")
print("   âœ… Zeile 2: if book.isbn in self.books (add_book)")
print("   âŒ Zeile 3: raise BookAlreadyExistsError (NICHT getestet!)")
print("   âœ… Zeile 4: self.books[book.isbn] = book")
print("   âœ… Zeile 5: if isbn not in self.books (get_book)")
print("   âŒ Zeile 6: raise BookNotFoundError (NICHT getestet!)")
print("   âœ… Zeile 7: return self.books[isbn]")
print("   âœ… Zeile 8: return len(self.books)")
print("   âŒ Zeile 9: if isbn not in self.books (remove_book)")
print("   âŒ Zeile 10: raise BookNotFoundError (NICHT getestet!)")
print("   âŒ Zeile 11: del self.books[isbn] (NICHT getestet!)")
print("\n" + "="*60)
print("ERGEBNIS: 6 von 11 Zeilen getestet = 55% Coverage")
print("="*60)
print("\nğŸ”´ Problem: 5 Zeilen werden NICHT getestet!")
print("   - Fehlerbehandlung (BookAlreadyExistsError)")
print("   - Fehlerbehandlung (BookNotFoundError)")
print("   - remove_book() Funktion komplett ungetestet!")
````

## 7. Praktische Tipps fÃ¼r gute Coverage

### Tipp 1: Testen Sie alle Pfade

Jede `if`-Anweisung hat **zwei Pfade**:

```python

    # Code A
else:              # â† Pfad 2: bedingung ist False
    # Code B
```

**Sie brauchen mindestens 2 Tests:**
- Test 1: bedingung ist True â†’ Code A wird ausgefÃ¼hrt
- Test 2: bedingung ist False â†’ Code B wird ausgefÃ¼hrt

---

### Tipp 2: Testen Sie alle Exceptions

Jede `raise`-Anweisung muss getestet werden:

```python

    raise MeinFehler("Etwas ist schiefgegangen!")
```

**Sie brauchen einen Test, der den Fehler auslÃ¶st!**

```python

    with pytest.raises(MeinFehler):
        # Code, der den Fehler auslÃ¶st
```

---

### Tipp 3: Testen Sie alle Funktionen

Jede Funktion/Methode sollte mindestens einen Test haben:

