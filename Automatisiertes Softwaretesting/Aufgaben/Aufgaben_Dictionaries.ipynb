```python
#Aufgabe 1
# Beispielslösung
personen = {
    "Anna": 25,
    "Ben": 30,
    "Clara": 22
}
print(personen)
```

```python
#Aufgabe 7
# Lösung
for name in personen:
    print(name)
```

```python
#Aufgabe 8
for alter in personen.values():
    print(alter)
```

```python
#Aufgabe 9
for name, alter in personen.items():
    print("Name:", name, ", Alter:", alter)
```

```python
#Aufgabe 10
personen.update({"Eva": 27})
print(personen)
```

```python
#Aufgabe 11
schluessel = list(personen.keys())
werte = list(personen.values())
print(schluessel) 
print(werte)  
```

```python
#Aufgabe 12
schluessel = ['Anna', 'Ben', 'David', 'Eva']
werte = [26, 30, 28, 27]
neues_dict = dict(zip(schluessel, werte))
print(neues_dict)
```
