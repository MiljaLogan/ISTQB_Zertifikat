## LÃ¶sung Bonusaufgabe: BMI-Rechner mit Bewertung

````python
def berechne_bmi(gewicht, grÃ¶ÃŸe):
    # BMI berechnen
    bmi = gewicht / (grÃ¶ÃŸe ** 2)
    
    # Bewertung
    if bmi < 18.5:
        bewertung = "Untergewicht"
    elif bmi < 25:
        bewertung = "Normalgewicht"
    elif bmi < 30:
        bewertung = "Ãœbergewicht"
    else:
        bewertung = "Adipositas"
    
    return bmi, bewertung

# Testen
bmi1, bewertung1 = berechne_bmi(70, 1.75)
print(f"BMI: {bmi1:.1f} - {bewertung1}")

# Weitere Tests:
print("\n--- Verschiedene BMI-Berechnungen ---")
testfÃ¤lle = [
    (55, 1.70),   # Untergewicht
    (70, 1.75),   # Normalgewicht
    (85, 1.75),   # Ãœbergewicht
    (100, 1.75),  # Adipositas
]

for gewicht, grÃ¶ÃŸe in testfÃ¤lle:
    bmi, bewertung = berechne_bmi(gewicht, grÃ¶ÃŸe)
    print(f"Gewicht: {gewicht}kg, GrÃ¶ÃŸe: {grÃ¶ÃŸe}m â†’ BMI: {bmi:.1f} ({bewertung})")
````
