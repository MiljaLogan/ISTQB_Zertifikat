```python
# Schritt 1: RED - Tests schreiben
# (Diese wÃ¼rden fehlschlagen, wenn die Funktion noch nicht existiert)

# Schritt 2: GREEN - Erste Implementation
def note_zu_text(note):
    if note == 1:
        return "Sehr gut"
    elif note == 2:
        return "Gut"
    elif note == 3:
        return "Befriedigend"
    elif note == 4:
        return "Ausreichend"
    elif note == 5:
        return "Mangelhaft"
    elif note == 6:
        return "UngenÃ¼gend"

# Tests
assert note_zu_text(1) == "Sehr gut"
assert note_zu_text(2) == "Gut"
assert note_zu_text(3) == "Befriedigend"
assert note_zu_text(4) == "Ausreichend"
assert note_zu_text(5) == "Mangelhaft"
assert note_zu_text(6) == "UngenÃ¼gend"
print("âœ“ Alle Tests bestanden!")

# Schritt 3: REFACTOR - Verbesserte Version mit Dictionary
def note_zu_text_refactored(note):
    noten = {
        1: "Sehr gut",
        2: "Gut",
        3: "Befriedigend",
        4: "Ausreichend",
        5: "Mangelhaft",
        6: "UngenÃ¼gend"
    }
    return noten[note]

# Tests fÃ¼r refactored Version
assert note_zu_text_refactored(1) == "Sehr gut"
assert note_zu_text_refactored(2) == "Gut"
assert note_zu_text_refactored(3) == "Befriedigend"
assert note_zu_text_refactored(4) == "Ausreichend"
assert note_zu_text_refactored(5) == "Mangelhaft"
assert note_zu_text_refactored(6) == "UngenÃ¼gend"
print("âœ“ Alle Tests nach Refactoring bestanden!")
print("\nRefactoring erfolgreich: Code ist jetzt kÃ¼rzer und wartbarer!")
```

## LÃ¶sung Bonusaufgabe: Komplexere Logik

```python
# Ã„quivalenzklassen:
# Gewicht: 0-1 kg, 1-5 kg, Ã¼ber 5 kg
# Versandart: Normal, Express
# â†’ 6 Kombinationen insgesamt

# Grenzwerte: 1 kg, 5 kg

def versandkosten(gewicht, express=False):
    # Basis-Versandkosten berechnen
    if gewicht <= 1:
        basis = 3
    elif gewicht <= 5:
        basis = 5
    else:
        basis = 8
    
    # Express-Aufschlag
    if express:
        return basis + 5
    else:
        return basis

# Tests fÃ¼r normale Lieferung
print("=== Tests fÃ¼r normale Lieferung ===")
assert versandkosten(0.5, False) == 3
print("âœ“ 0.5 kg normal = 3 â‚¬")

assert versandkosten(1, False) == 3  # Grenzwert
print("âœ“ 1 kg normal = 3 â‚¬ (Grenzwert)")

assert versandkosten(3, False) == 5
print("âœ“ 3 kg normal = 5 â‚¬")

assert versandkosten(5, False) == 5  # Grenzwert
print("âœ“ 5 kg normal = 5 â‚¬ (Grenzwert)")

assert versandkosten(6, False) == 8
print("âœ“ 6 kg normal = 8 â‚¬")

# Tests fÃ¼r Express-Lieferung
print("\n=== Tests fÃ¼r Express-Lieferung ===")
assert versandkosten(0.5, True) == 8
print("âœ“ 0.5 kg express = 8 â‚¬ (3 + 5)")

assert versandkosten(1, True) == 8  # Grenzwert
print("âœ“ 1 kg express = 8 â‚¬ (Grenzwert)")

assert versandkosten(3, True) == 10
print("âœ“ 3 kg express = 10 â‚¬ (5 + 5)")

assert versandkosten(5, True) == 10  # Grenzwert
print("âœ“ 5 kg express = 10 â‚¬ (Grenzwert)")

assert versandkosten(6, True) == 13
print("âœ“ 6 kg express = 13 â‚¬ (8 + 5)")

print("\nâœ“ Alle Tests fÃ¼r Bonusaufgabe bestanden!")
```
