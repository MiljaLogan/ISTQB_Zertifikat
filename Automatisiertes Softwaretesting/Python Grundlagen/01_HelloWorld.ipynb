**Das "Hallo Welt!"-Programm schreiben**

Es ist eine weit verbreitete Tradition, das Programmieren mit einem einfachen Programm zu beginnen, das den Text "Hallo Welt!" ausgibt. Lass uns das tun!

Gib folgenden Code in deine gerade erstellte Datei `erstes_programm.py` ein:


````python
print('Hallo Welt!')
````



**Was bedeutet dieser Code? Eine Analyse**

Schauen wir uns die Codezeile genauer an:

1.  **`print`**: Das ist der Name einer eingebauten **Funktion** in Python.
    *   **Was sind Funktionen?** Funktionen sind wiederverwendbare Codeblöcke, die eine bestimmte Aufgabe erledigen. Stell dir vor, sie sind wie Werkzeuge: Ein Hammer ist für Nägel da, eine Säge zum Schneiden. Die `print`-Funktion ist dafür da, etwas auf dem Bildschirm anzuzeigen. Es gibt viele andere Funktionen in Python, z.B. um Berechnungen durchzuführen oder Daten zu sortieren.
2.  **`()`**: Die runden Klammern direkt nach dem Funktionsnamen `print` sind notwendig, um eine Funktion *aufzurufen* (also zu benutzen).
3.  **`'Hallo Welt!'`**: Das ist der Text, den wir ausgeben möchten.
    *   **Argument:** Die Information, die wir einer Funktion innerhalb der Klammern übergeben, nennt man **Argument**. Hier ist `'Hallo Welt!'` das Argument für die `print`-Funktion. Wir teilen ihr damit mit, *was* sie ausgeben soll.
    *   **String:** In der Programmierung wird Text wie `'Hallo Welt!'` als **String** (Zeichenkette) bezeichnet. Strings sind ein grundlegender **Datentyp** in Python. Andere Datentypen sind zum Beispiel Zahlen (Integer für ganze Zahlen, Float für Kommazahlen), die wir später kennenlernen werden.
    *   **Anführungszeichen:** Beachte die einfachen Anführungszeichen (`'`) um den Text. In Python müssen Strings immer von einfachen (`'`) oder doppelten (`"`) Anführungszeichen umschlossen sein. Diese Zeichen zeigen Python, wo der String beginnt und endet. Sie selbst sind aber *nicht* Teil des Textes, der ausgegeben wird. Es ist eine gute Praxis, konsistent zu sein – wir werden uns hier meistens an einfache Anführungszeichen halten.

Zusammenfassend: Mit der Zeile `print('Hallo Welt!')` *rufen* wir die `print`-Funktion *auf* und *übergeben* ihr den *String* `'Hallo Welt!'` als *Argument*, damit dieser auf dem Bildschirm ausgegeben wird.
