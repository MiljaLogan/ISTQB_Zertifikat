**Ziel:** In dieser Lerneinheit lernst du die grundlegenden mathematischen Operatoren in Python kennen. Du erfährst, wie du Addition, Subtraktion, Multiplikation und verschiedene Arten der Division in deinem Code durchführen kannst.



**1. Was sind Operatoren?**

Operatoren sind spezielle Symbole in Python (und vielen anderen Programmiersprachen), die verwendet werden, um Operationen mit Werten (sogenannten Operanden) durchzuführen. Die bekanntesten sind die mathematischen Operatoren.

**2. Grundlegende Arithmetik**

*   **Addition (`+`):** Addiert zwei Zahlen.



````python
print(2 + 3) # Ausgabe: 5 
````




*   **Subtraktion (`-`):** Subtrahiert die zweite Zahl von der ersten.



````python
print(5 - 4) # Ausgabe: 1
````




*   **Multiplikation (`*`):** Multipliziert zwei Zahlen. Das Symbol ist das Sternchen.



````python
print(3 * 5) # Ausgabe: 15
````




**3. Die drei Arten der Division**

Für die Division bietet Python gleich drei verschiedene Operatoren mit unterschiedlichen Ergebnissen:

*   **Standard-Division (`/`):** Führt eine normale Division durch und gibt das Ergebnis **immer** als Fließkommazahl (`float`) zurück, auch wenn die Division "glatt aufgeht".



````python
print(6 / 2) # Ausgabe: 3.0 
print(7 / 2) # Ausgabe: 3.5
````




*   **Ganzzahl-Division / Floor Division (`//`):** Führt eine Division durch und gibt nur den ganzzahligen Anteil des Ergebnisses zurück, wobei **immer abgerundet** wird (zur nächstkleineren ganzen Zahl).
    ```python


````python
print(6 // 2) # Ausgabe: 3
print(7 // 2) # Ausgabe: 3 (3.5 abgerundet ist 3)
print(7.5 // 2) # Ausgabe: 3.0 (Auch hier wird abgerundet, Ergebnis ist float wenn ein Operand float ist)
````



*   **Modulo-Operator (`%`):** Gibt den **Rest** einer Division zurück.
    ```python


````python
print(6 % 2) # Ausgabe: 0 (6 geteilt durch 2 ist 3 ohne Rest)
print(7 % 2) # Ausgabe: 1 (7 geteilt durch 2 ist 3, Rest 1)
````


    Der Modulo-Operator ist nützlich, um z.B. zu prüfen, ob eine Zahl gerade (`zahl % 2 == 0`) oder ungerade ist.

**Wichtige Regel: Division durch Null**
Du darfst in der Mathematik und auch in Python **niemals** durch Null teilen. Der Versuch, dies mit `/`, `//` oder `%` zu tun, führt zu einem Fehler und dein Programm stürzt ab!

```python
# print(5 / 0) # Führt zu einem ZeroDivisionError!
# print(5 // 0) # Führt zu einem ZeroDivisionError!
# print(5 % 0) # Führt zu einem ZeroDivisionError!
```

**4. Potenzierung (Exponentiation)**

*   **Potenz-Operator (`**`):** Erhebt die erste Zahl zur Potenz der zweiten Zahl (Basis hoch Exponent). Das Symbol sind zwei Sternchen.


````python
print(2 ** 3) # Ausgabe: 8 (entspricht 2 * 2 * 2)
print(3 ** 2) # Ausgabe: 9 (entspricht 3 * 3)
````



**5. Datentypen der Ergebnisse**

Der Datentyp des Ergebnisses einer Operation hängt von den Datentypen der Operanden ab:

*   Wenn du `+`, `-`, `*`, `//`, `%` oder `**` mit **zwei Integers** verwendest, ist das Ergebnis normalerweise ein **Integer**.


````python
print(type(10 + 5))   # <class 'int'>
print(type(10 // 3))  # <class 'int'>
print(type(2 ** 4))   # <class 'int'>
````



*   Wenn mindestens einer der Operanden ein **Float** ist, ist das Ergebnis bei `+`, `-`, `*`, `//`, `%`, `**` ebenfalls ein **Float**.


````python
print(type(10.0 + 5)) # <class 'float'>
print(type(10 * 2.5)) # <class 'float'>
print(type(7.0 // 2)) # <class 'float'>
print(type(2.0 ** 3)) # <class 'float'>
````



*   Die **Standard-Division (`/`)** ergibt **immer** einen **Float**, unabhängig von den Operanden.



````python
print(type(6 / 2))   # <class 'float'> (obwohl das Ergebnis 3.0 ist)
print(type(7 / 2))   # <class 'float'>
````


    ```

**Zusammenfassung**

*   Die grundlegenden mathematischen Operatoren sind `+`, `-`, `*`.
*   Für die Division gibt es `/` (Standard, Ergebnis immer float), `//` (Ganzzahl, abgerundet) und `%` (Rest).
*   Division durch Null ist verboten und führt zu einem `ZeroDivisionError`.
*   Potenzierung erfolgt mit `**`.
*   Der Datentyp des Ergebnisses hängt von den Operatoren und Operanden ab (insbesondere `/` ergibt immer `float`).
