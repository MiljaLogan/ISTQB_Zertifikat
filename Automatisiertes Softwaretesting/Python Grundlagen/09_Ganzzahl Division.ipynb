**Zusatz: Achtung bei der Ganzzahl-Division (`//`) und dem Ergebnistyp!**

Über Operatoren haben wir gelernt, dass die Ganzzahl-Division (`//`) das Ergebnis der Division immer zur nächstkleineren ganzen Zahl **abrundet**.

Es ist jedoch wichtig zu beachten, dass diese "ganze Zahl" als Ergebnis entweder vom Typ **Integer (`int`)** oder vom Typ **Float (`float`)** sein kann, abhängig von den Datentypen der Zahlen, die du dividierst!

Schau dir diese Beispiele an:


```python
print(6 // 3)     # Ergibt 2   (Typ: int) - Weil beide Zahlen (6 und 3) Integer sind.
```

```python
print(6.0 // 3)   # Ergibt 2.0 (Typ: float) - Weil mindestens eine Zahl (6.0) ein Float ist.
```

```python
print(6 // 3.0)   # Ergibt 2.0 (Typ: float) - Weil mindestens eine Zahl (3.0) ein Float ist.
```

```python
print(6.0 // 3.0) # Ergibt 2.0 (Typ: float) - Weil beide Zahlen Floats sind.
```



**Die Regel lautet:**

*   Wenn du zwei **Integer (`int`)** mit `//` dividierst, ist das Ergebnis ein **Integer (`int`)**.
*   Wenn aber **mindestens eine** der beteiligten Zahlen ein **Float (`float`)** ist, dann ist das Ergebnis der `//`-Operation ebenfalls ein **Float (`float`)**, auch wenn der Wert nach dem Abrunden zufällig keine Nachkommastellen hat (wie z.B. `2.0`).

Dieselbe Regel (mindestens ein Float führt zu einem Float-Ergebnis) gilt auch für die anderen Operatoren (`+`, `-`, `*`, `%`, `**`). Die einzige Ausnahme ist die **Standard-Division (`/`)**, die **immer** einen **Float** zurückgibt, selbst wenn du zwei Integer dividierst (z.B. `6 / 3` ergibt `2.0`).

Das kann manchmal etwas knifflig sein, also behalte diesen Unterschied im Hinterkopf, besonders wenn der genaue Datentyp des Ergebnisses wichtig ist (z.B. in Prüfungen oder bei späteren komplexeren Berechnungen).
