
**Lerneinheit: Kommentare in Python – Notizen im Code**

**Ziel:** In dieser Lerneinheit lernst du, wie du Kommentare in deinem Python-Code hinterlassen kannst. Du erfährst, was Kommentare sind, wie man sie erstellt und warum sie ein wichtiges Werkzeug beim Programmieren sind.



**1. Was sind Kommentare?**

Kommentare sind Teile deines Python-Codes, die vom Python-Interpreter **ignoriert** werden. Das bedeutet, sie haben absolut keinen Einfluss darauf, wie dein Programm ausgeführt wird. Sie sind reine Notizen für Menschen – also für dich selbst oder für andere Personen, die deinen Code lesen.

**2. Wie erstellt man Kommentare?**

In Python beginnt ein Kommentar mit dem **Rautezeichen (`#`)**. Alles, was in einer Zeile *nach* dem `#` steht, wird als Kommentar behandelt und nicht ausgeführt.

**Beispiel:**



```python
# Dies ist ein Kommentar, der eine ganze Zeile einnimmt.
print("Hallo Welt!") # Dieser Code wird ausgeführt.

print("Noch eine Ausgabe.") # Dies ist ein Kommentar am Ende der Zeile.
```


Führe diesen Code aus. Du wirst sehen, dass nur die `print`-Anweisungen eine Ausgabe erzeugen. Die Texte nach den `#`-Zeichen werden ignoriert.

**3. Wozu braucht man Kommentare?**

Kommentare sind sehr nützlich und haben hauptsächlich zwei Anwendungszwecke:

*   **a) Code erklären und dokumentieren:**
    Wenn dein Code komplexer wird, ist es oft hilfreich, kurze Erklärungen hinzuzufügen. Das hilft nicht nur anderen, deinen Code zu verstehen, sondern auch dir selbst, wenn du nach einiger Zeit auf deinen eigenen Code zurückblickst und dich fragst, warum du etwas auf eine bestimmte Art und Weise gemacht hast.



```python
# Wir speichern die Benutzer-ID als String, 
# da sie führende Nullen haben könnte und wir nicht damit rechnen.
benutzer_id = '007' 

# Berechnung des Rabatts (10%)
preis = 100
rabatt = preis * 0.10 # 10% Rabatt
endpreis = preis - rabatt

print(endpreis)
```



*   **b) Code vorübergehend deaktivieren ("Auskommentieren"):**
    Manchmal möchtest du einen Teil deines Codes temporär deaktivieren, vielleicht um etwas zu testen oder weil du ihn gerade nicht brauchst, aber auch nicht löschen willst. Anstatt den Code zu löschen, kannst du ihn einfach *auskommentieren*, indem du ein `#` an den Anfang jeder Zeile setzt.



```python
print("Dieser Teil wird ausgeführt.")
    
# Der folgende Block ist vorübergehend deaktiviert:
# print("Diese Zeile wird NICHT ausgeführt.")
# alter = 30
# print(alter)

print("Dieser Teil wird wieder ausgeführt.") 
```


    Wenn du den Code später wieder aktivieren möchtest, entfernst du einfach die `#`-Zeichen am Anfang der entsprechenden Zeilen.

**Zusammenfassung**

*   Kommentare beginnen mit `#`.
*   Alles nach dem `#` in einer Zeile wird von Python ignoriert.
*   Kommentare dienen dazu, Code zu erklären oder Teile davon temporär zu deaktivieren.
*   Gute Kommentare machen deinen Code verständlicher und wartbarer.
