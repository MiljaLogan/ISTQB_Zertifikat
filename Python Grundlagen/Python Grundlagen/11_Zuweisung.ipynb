**Lerneinheit: Werte neu zuweisen und Operatoren mit Strings**

**Ziel:** In dieser Lerneinheit vertiefst du dein Verständnis des Zuweisungsoperators (`=`), lernst verkürzte Schreibweisen für Operationen und Zuweisungen kennen und erfährst, wie die Operatoren `+` und `*` mit Strings funktionieren.


**1. Rückblick: Zuweisungsoperator (`=`)**

Du kennst bereits den Zuweisungsoperator (`=`), um einer Variable einen Wert zuzuweisen:



```python
alter = 28 
# Erstellt die Variable 'alter' und speichert den Integer-Wert 28 darin.
print(alter) 
```



**2. Zuweisung mit Berechnung**

Du kannst Operatoren auch auf der rechten Seite der Zuweisung verwenden. Python berechnet zuerst das Ergebnis des Ausdrucks rechts vom `=` und weist dieses Ergebnis dann der Variable links vom `=` zu.



```python
alter = 28
neues_alter = alter + 5 
# Zuerst wird alter + 5 (28 + 5 = 33) berechnet.
# Dann wird das Ergebnis (33) der Variable neues_alter zugewiesen.
print(neues_alter) # Ausgabe: 33
```



**3. Den Wert einer Variable basierend auf sich selbst ändern**

Es ist sehr üblich, den Wert einer Variable zu nehmen, eine Operation durchzuführen und das Ergebnis wieder in derselben Variable zu speichern.



```python
alter = 28
print("Ursprüngliches Alter:", alter)
```

```python
alter = alter + 7 
# 1. Nimm den aktuellen Wert von alter (28).
# 2. Addiere 7 hinzu (28 + 7 = 35).
# 3. Speichere das neue Ergebnis (35) zurück in der Variable alter. 
#    Der alte Wert (28) wird überschrieben.

print("Alter nach Addition:", alter) # Ausgabe: Alter nach Addition: 35
```



**4. Verkürzte Zuweisungsoperatoren (Augmented Assignment)**

Für die häufige Operation "nimm den aktuellen Wert, führe eine Operation aus, speichere zurück" gibt es praktische Abkürzungen:

*   **`+=` (Additionszuweisung):** `x += y` ist die Kurzform für `x = x + y`.
*   **`-=` (Subtraktionszuweisung):** `x -= y` ist die Kurzform für `x = x - y`.
*   **`*=` (Multiplikationszuweisung):** `x *= y` ist die Kurzform für `x = x * y`.
*   **`/=` (Divisionszuweisung):** `x /= y` ist die Kurzform für `x = x / y`.
*   **`//=` (Ganzzahl-Divisionszuweisung):** `x //= y` ist die Kurzform für `x = x // y`.
*   **`%=` (Modulo-Zuweisung):** `x %= y` ist die Kurzform für `x = x % y`.
*   **`**=` (Potenzierungszuweisung):** `x **= y` ist die Kurzform für `x = x ** y`.

**Beispiele:**


```python
alter = 28

alter += 7  # Entspricht: alter = alter + 7
print(f"Nach += 7: {alter}") # f-String für formatierte Ausgabe

alter *= 2  # Entspricht: alter = alter * 2
print(f"Nach *= 2: {alter}") 

alter -= 5  # Entspricht: alter = alter - 5
print(f"Nach -= 5: {alter}") 

alter /= 3  # Entspricht: alter = alter / 3 (Achtung: Ergebnis wird Float!)
print(f"Nach /= 3: {alter}") 
```



**5. Operatoren mit Strings**

Die Operatoren `+` und `*` haben eine spezielle Bedeutung, wenn sie mit Strings verwendet werden:

*   **`+` (String-Verkettung / Konkatenation):** Verbindet zwei oder mehr Strings zu einem einzigen String.


```python
teil1 = "Hokus"
teil2 = "pokus"
magie = teil1 + teil2 
print(magie) # Ausgabe: Hokuspokus 
```

```python
# Beachte: Python fügt KEIN Leerzeichen automatisch hinzu!
vorname = "Max"
nachname = "Mustermann"
voller_name = vorname + " " + nachname # Leerzeichen muss explizit hinzugefügt werden
print(voller_name) # Ausgabe: Max Mustermann
```


    Das Zusammenfügen von Strings wird in der IT oft als **Verkettung** oder **Konkatenation** bezeichnet.

*   **`*` (String-Wiederholung):** Wiederholt einen String so oft, wie durch eine ganze Zahl angegeben.


```python
wort = "Echo! "
wiederholung = wort * 5
print(wiederholung) # Ausgabe: Echo! Echo! Echo! Echo! Echo! 
```



**6. Achtung: Datentypen sind entscheidend!**

Da der `+` Operator sowohl für die Addition von Zahlen als auch für die Verkettung von Strings verwendet wird, musst du genau auf die Datentypen achten:



```python
# Beispiel 1: Addition von Zahlen (Integers)
zahl1 = 2
zahl2 = 8
summe = zahl1 + zahl2
print(f"{zahl1} + {zahl2} = {summe} (Typ: {type(summe)})") 
# Ausgabe: 2 + 8 = 10 (Typ: <class 'int'>)
```

```python
# Beispiel 2: Verkettung von Strings
text1 = "2"
text2 = "8"
verkettet = text1 + text2
print(f"'{text1}' + '{text2}' = '{verkettet}' (Typ: {type(verkettet)})")
# Ausgabe: '2' + '8' = '28' (Typ: <class 'str'>) 
```



Im zweiten Fall werden die Zeichen '2' und '8' einfach hintereinander gehängt, da es sich um Strings handelt – es findet keine mathematische Addition statt!

**Zusammenfassung**

*   Der Zuweisungsoperator (`=`) speichert das Ergebnis der rechten Seite in der Variable auf der linken Seite.
*   Variablen können basierend auf ihrem eigenen Wert neu zugewiesen werden (`x = x + 1`).
*   Verkürzte Zuweisungsoperatoren (`+=`, `-=`, `*=`, `/=` etc.) vereinfachen diese Neuzuweisungen.
*   Der `+` Operator verkettet Strings (fügt sie aneinander).
*   Der `*` Operator wiederholt einen String, wenn er mit einer Ganzzahl verwendet wird.
*   Der Effekt des `+` Operators hängt stark von den Datentypen der Operanden ab (Addition bei Zahlen, Verkettung bei Strings).
