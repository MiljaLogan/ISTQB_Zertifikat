
## Szenario 1: BMI-Rechner für Fitnessstudio

**Ausgangssituation:**
Ein lokales Fitnessstudio möchte seinen Mitgliedern einen einfachen BMI-Rechner zur Verfügung stellen. Der Rechner soll den Body-Mass-Index berechnen und eine einfache Bewertung ausgeben.

**Anforderungen:**
- Das Programm nimmt Gewicht in kg (float) und Größe in Metern (float) entgegen
- Berechnung: BMI = Gewicht / (Größe²)
- Ausgabe des berechneten BMI-Werts (gerundet auf 1 Dezimalstelle)
- Kategorisierung:
  - BMI < 18.5: "Untergewicht"
  - BMI 18.5 - 24.9: "Normalgewicht"
  - BMI 25.0 - 29.9: "Übergewicht"
  - BMI ≥ 30.0: "Adipositas"
- Eingabevalidierung: Gewicht > 0 und Größe > 0

**Aufgaben:**
1. Entwerft mindestens 5 Testfälle (normale Werte, Grenzwerte, ungültige Eingaben)
2. Schreibt die Tests in Python und führt sie mit pytest aus
3. Implementiert den BMI-Rechner

## Szenario 2: Notenrechner für Prüfungen

**Ausgangssituation:**
Eine Berufsschule benötigt ein Programm zur Umrechnung von Punkten in Noten nach dem IHK-Schema.

**Anforderungen:**
- Das Programm nimmt erreichte Punkte und maximale Punktzahl entgegen (beide als int)
- Berechnung des Prozentsatzes: (erreichte Punkte / maximale Punkte) × 100
- Notenausgabe nach folgendem Schema:
  - 92-100%: Note 1 (sehr gut)
  - 81-91%: Note 2 (gut)
  - 67-80%: Note 3 (befriedigend)
  - 50-66%: Note 4 (ausreichend)
  - 30-49%: Note 5 (mangelhaft)
  - 0-29%: Note 6 (ungenügend)
- Eingabevalidierung: Beide Werte müssen ≥ 0 sein, erreichte Punkte ≤ maximale Punkte

**Aufgaben:**
1. Entwerft mindestens 7 Testfälle (für jede Note + Grenzwerte + Fehlerfälle)
2. Schreibt die Tests in Python und führt sie mit pytest aus
3. Implementiert den Notenrechner

## Szenario 3: Rabattrechner für Online-Shop

**Ausgangssituation:**
Ein Online-Shop möchte ein automatisches Rabattsystem einführen, das je nach Bestellwert unterschiedliche Rabatte gewährt.

**Anforderungen:**
- Das Programm nimmt den Warenwert (float) entgegen
- Rabattstaffel:
  - unter 50€: kein Rabatt (0%)
  - 50€ bis 99.99€: 5% Rabatt
  - 100€ bis 199.99€: 10% Rabatt
  - ab 200€: 15% Rabatt
- Ausgabe: Ursprungspreis, Rabatt in €, Endpreis (jeweils auf 2 Dezimalstellen gerundet)
- Eingabevalidierung: Warenwert muss ≥ 0 sein

**Aufgaben:**
1. Entwerft mindestens 6 Testfälle (Grenzwerte für jede Rabattstufe + Fehlerfälle)
2. Schreibt die Tests in Python und führt sie mit pytest aus
3. Implementiert den Rabattrechner
