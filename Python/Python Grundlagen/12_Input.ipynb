**Lerneinheit: Benutzereingaben mit `input()`**

**Ziel:** Bisher konnten unsere Programme nur vordefinierte Informationen ausgeben. In dieser Lerneinheit lernst du die `input()`-Funktion kennen, mit der du dein Programm interaktiv gestalten kannst, indem du den Benutzer zur Eingabe von Informationen aufforderst.


**1. Programme interaktiv machen: Die `input()`-Funktion**

Neben der `print()`-Funktion, die Daten *ausgibt*, gibt es die `input()`-Funktion, die Daten vom Benutzer *einliest*.

*   **Was `input()` tut:**
    1.  Das Programm hält an und wartet auf den Benutzer.
    2.  Der Benutzer kann Text in die Konsole/das Terminal eingeben und mit der `Enter`-Taste bestätigen.
    3.  Die `input()`-Funktion nimmt diesen eingegebenen Text entgegen und gibt ihn als **Wert** an das Programm zurück.

**2. Ein einfaches Beispiel**

Lass uns ein Programm schreiben, das den Benutzer nach seinem Namen fragt und ihn dann begrüßt:



```python
# 1. Den Benutzer zur Eingabe auffordern (separate print-Anweisung)
print(()"Wie ist dein Name?")

# 2. Auf die Eingabe warten und das Ergebnis in einer Variable speichern
benutzername = input() 

# 3. Den gespeicherten Namen in einer Begrüßung verwenden
print("Hallo,", benutzername) 
```



**Führe diesen Code aus:**

1.  Im Terminal erscheint die Frage "Wie ist dein Name?".
2.  Darunter blinkt wahrscheinlich ein Cursor. Das Programm wartet jetzt!
3.  Gib deinen Namen ein (z.B. "Maria") und drücke `Enter`.
4.  Das Programm läuft weiter und gibt die Begrüßung aus: "Hallo, Maria".

**3. Wie Funktionen arbeiten: Effekt vs. Rückgabewert**

Generell können Funktionen in Python zwei Hauptaufgaben haben:

*   **Einen Effekt auslösen:** Sie tun etwas Sichtbares oder Hörbares, z.B. Text anzeigen (`print`), eine Datei speichern, einen Ton abspielen.
*   **Einen Wert zurückgeben:** Sie berechnen oder holen einen Wert und liefern ihn an die Stelle im Code zurück, wo die Funktion aufgerufen wurde. Diesen zurückgegebenen Wert können wir dann oft in einer Variable speichern.

Vergleichen wir `print` und `input`:

*   `print()`: Hat einen **Effekt** (zeigt Text an), gibt aber keinen wirklich nützlichen Wert zurück, den wir speichern müssten.
*   `input()`: Hat einen **Effekt** (zeigt optional eine Eingabeaufforderung an, wartet auf den Benutzer) UND gibt einen **Wert zurück** (nämlich das, was der Benutzer eingegeben hat).

Weil `input()` einen Wert zurückgibt, ist es sinnvoll, diesen Wert mit dem Zuweisungsoperator (`=`) in einer Variable zu speichern (wie in `benutzername = input()`), um ihn später im Programm verwenden zu können.

**4. Die Eingabeaufforderung direkt in `input()` integrieren**

Es ist üblich, dem Benutzer direkt mitzuteilen, was er eingeben soll. Anstatt zuerst `print()` und dann `input()` zu verwenden, können wir den Text für die Eingabeaufforderung (den "Prompt") direkt als Argument an die `input()`-Funktion übergeben:


```python
# Kombinierte Eingabeaufforderung und Warten auf Eingabe
benutzername = input("Wie ist dein Name? ") # Leerzeichen am Ende ist benutzerfreundlich

# Den gespeicherten Namen verwenden
print("Hallo,", benutzername)
```



Führe diesen Code aus. Er funktioniert genauso wie der vorherige, ist aber kürzer. Der Text "Wie ist dein Name? " wird angezeigt, und das Programm wartet direkt dahinter auf deine Eingabe.

**5. Wichtiger Hinweis: `input()` gibt immer einen String zurück!**

Das ist ein sehr wichtiger Punkt: Unabhängig davon, was der Benutzer eingibt – auch wenn er nur Ziffern eingibt –, gibt die `input()`-Funktion das Ergebnis **immer** als **String (`str`)** zurück.

**Beispiel:**



```python
alter_eingabe = input("Wie alt bist du? ")
print(f"Der Typ der Eingabe ist: {type(alter_eingabe)}") 
```



Wenn du hier z.B. `30` eingibst, wird die Ausgabe sein:
`Der Typ der Eingabe ist: <class 'str'>`

Das bedeutet, du kannst mit dem Ergebnis von `input()` nicht sofort mathematische Berechnungen durchführen, selbst wenn es wie eine Zahl aussieht. Wie wir Text-Eingaben in Zahlen umwandeln können, um damit zu rechnen, lernen wir in einer späteren Lektion.

**Zusammenfassung**

*   Die `input()`-Funktion hält das Programm an und wartet auf Benutzereingaben im Terminal.
*   Sie gibt den eingegebenen Text als **String (`str`)** zurück.
*   Der zurückgegebene Wert wird üblicherweise in einer Variable gespeichert.
*   Man kann eine Eingabeaufforderung (Prompt) als Argument an `input()` übergeben (`input("Deine Frage: ")`).
*   Achtung: Der Rückgabewert ist *immer* ein String, auch bei Zahleneingaben.
