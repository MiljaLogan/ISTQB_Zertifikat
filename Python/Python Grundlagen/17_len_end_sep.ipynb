**1. Die Länge eines Strings ermitteln: `len()`**

Oft ist es nützlich zu wissen, wie viele Zeichen ein String enthält. Dafür gibt es in Python die eingebaute Funktion `len()`. Sie nimmt einen String (oder andere sequenzielle Datentypen) als Argument und gibt dessen Länge als ganze Zahl zurück.



```python
mein_text = 'Hallo!'
laenge = len(mein_text)
print(f"Der String '{mein_text}' hat {laenge} Zeichen.") # Ausgabe: Der String 'Hallo!' hat 6 Zeichen.
```

```python
print(len("Python ist toll")) # Gibt direkt die Länge aus: 15
```

```python
print(len("")) # Ein leerer String hat die Länge 0
```



**2. `print()` und der Zeilenumbruch: Das `end`-Argument**

Du weißt bereits, dass jeder `print()`-Aufruf standardmäßig einen Zeilenumbruch (`\n`) am Ende der Ausgabe hinzufügt. Deshalb erscheinen mehrere `print()`-Anweisungen untereinander:



```python
print("Erste Zeile")
print("Zweite Zeile")
```



Dieses Verhalten kann man ändern! Die `print()`-Funktion hat optionale **Keyword-Argumente** (auch benannte Argumente genannt). Das sind Argumente, die du am Ende des Funktionsaufrufs mit `argumentname=wert` angibst.

Eines dieser Argumente ist `end`. Es legt fest, was *anstelle* des standardmäßigen Zeilenumbruchs am Ende der Ausgabe geschrieben wird. Der Standardwert von `end` ist `'\n'`.

**Beispiel:** Wir wollen keinen Zeilenumbruch, sondern einen Punkt und ein Leerzeichen:



```python
print("Das ist der erste Teil", end='. ')
print("Das ist der zweite Teil.") # Hier wird der Standard-Umbruch wieder verwendet
```

Wenn du gar nichts am Ende hinzufügen möchtest (auch keinen Umbruch), kannst du einen leeren String verwenden:



```python
print("Ohne Umbruch", end='')
print(" geht es hier weiter.")
```


**Keyword-Argumente (kurz erklärt):**
*   Sie werden nach den normalen (positionalen) Argumenten angegeben.
*   Sie werden über ihren Namen identifiziert (`end=...`, `sep=...`).
*   Sie sind immer optional, da sie Standardwerte haben. Wenn du sie nicht angibst, wird der Standardwert verwendet.

**3. Mehrere Elemente ausgeben: Das `sep`-Argument**

Du weißt auch, dass `print()` automatisch ein Leerzeichen zwischen mehreren Argumenten einfügt, die durch Komma getrennt sind:



```python
vorname = "Max"
nachname = "Mustermann"
print("Name:", vorname, nachname)
```


Dieses Trennzeichen (Separator) wird durch das Keyword-Argument `sep` gesteuert. Der Standardwert von `sep` ist ein Leerzeichen (`' '`). Du kannst diesen Wert überschreiben, um ein anderes Trennzeichen zu verwenden:

**Beispiel:** Trennen mit Bindestrichen:



```python
tag = "17"
monat = "07"
jahr = "2024"
print(tag, monat, jahr, sep='-')
```


**Beispiel:** Trennen mit ` | `:



```python
print("Apfel", "Birne", "Banane", sep=' | ')
```


**4. `sep` und `end` kombinieren**

Du kannst beide Keyword-Argumente, `sep` und `end`, im selben `print()`-Aufruf verwenden:



```python
items = ["Eins", "Zwei", "Drei"]
print(items[0], items[1], items[2], sep=', ', end='.\nENDE\n') # Items mit Komma+Leerzeichen trennen, am Ende Punkt+Umbruch+ENDE+Umbruch
```



**Zusammenfassung**

*   Mit `len(string)` erhältst du die Anzahl der Zeichen in einem String.
*   Die `print()`-Funktion hat optionale Keyword-Argumente wie `end` und `sep`.
*   `end` bestimmt, was am Ende der `print()`-Ausgabe steht (Standard: `'\n'` für Zeilenumbruch).
*   `sep` bestimmt das Trennzeichen zwischen mehreren Argumenten in `print()` (Standard: `' '` für Leerzeichen).
*   Du kannst `end` und `sep` nach Bedarf anpassen, um deine Ausgabe zu formatieren.
