 Refactoring (Code verbessern)

Jetzt schauen wir uns den Code an: KÃ¶nnen wir etwas verbessern?

Fragen fÃ¼r Refactoring:
- Ist der Code lesbar?
- Sind die Variablennamen verstÃ¤ndlich?
- Gibt es doppelten Code? 
- KÃ¶nnen wir etwas vereinfachen?

### Wichtig beim Refactoring:

- Nach jeder Ã„nderung: Test erneut ausfÃ¼hren!
- Test muss grÃ¼n bleiben âœ“
- Wenn Test rot wird â†’ Ã„nderung rÃ¼ckgÃ¤ngig machen

```python
# Alle Test-Funktionen

def test_addiere_zwei_zahlen():
    calc = Calculator()
    ergebnis = calc.addiere(2, 3)
    assert ergebnis == 5

def test_subtrahiere_zwei_zahlen():
    calc = Calculator()
    ergebnis = calc.subtrahiere(10, 4)
    assert ergebnis == 6

def test_dividiere_zwei_zahlen():
    calc = Calculator()
    ergebnis = calc.dividiere(10, 2)
    assert ergebnis == 5

def test_dividiere_durch_null_wirft_fehler():
    calc = Calculator()
    try:
        calc.dividiere(10, 0)
        assert False, "Es wurde kein Fehler geworfen!"
    except ValueError:
        pass  # Test bestanden!

# Alle Tests ausfÃ¼hren
tests = [
    ("test_addiere_zwei_zahlen", test_addiere_zwei_zahlen),
    ("test_subtrahiere_zwei_zahlen", test_subtrahiere_zwei_zahlen),
    ("test_dividiere_zwei_zahlen", test_dividiere_zwei_zahlen),
    ("test_dividiere_durch_null_wirft_fehler", test_dividiere_durch_null_wirft_fehler)
]

tests_bestanden = 0
tests_gesamt = len(tests)

print("FÃ¼hre Tests aus...\n")

for test_name, test_func in tests:
    try:
        test_func()
        print(f"âœ“ {test_name}")
        tests_bestanden += 1
    except AssertionError as e:
        print(f"âœ— {test_name}: {e}")
    except Exception as e:
        print(f"âœ— {test_name}: {e}")

print(f"\n{'='*50}")
print(f"{tests_bestanden}/{tests_gesamt} Tests bestanden")
print(f"{'='*50}")

if tests_bestanden == tests_gesamt:
    print("\nðŸŽ‰ Alle Tests sind grÃ¼n! Perfekt!")
```
